We first start off by analyzing the monthly usage patterns of the bicycles in London.

### 4.1 What is the average monthly bicycle usage?

```python
#studying average monthly bicycle usage for whole dataset
sql_query1 = """
SELECT
    MONTH(start_date) AS Month,
    COUNT(rental_id) AS Bicycle_Usage_Count
FROM cycle_hires
GROUP BY MONTH(start_date)
ORDER BY Month
"""
monthly_bicycle_usage = spark.sql(sql_query1)
monthly_bicycle_usage.show()
```

It can be observed from the table that June, July and August i.e, summer months are the most popular for bicycle usage. On the other hand, the winter months of December, January and February are the least popular for bicycle usage.

After studying the usage patterns, we move on to studying the ride duration trends and statistics.

### 4.2 What is the distribution and statistics of ride durations?

```python
#distribution of ride durations column
sql_query2 = """
SELECT
    COUNT(duration) AS count,
    AVG(duration) AS mean,
    STDDEV(duration) AS stddev,
    MIN(duration) AS min,
    MAX(duration) AS max
FROM cycle_hires
"""
duration_description = spark.sql(sql_query2)
duration_description.show()
```

Mean duration value is around 1307 seconds and the column has a high standard deviation value of around 7376 seconds. This indicates a wide variability in usage times. The presence of negative values suggests data inaccuracies, while the maximum duration of over 5.6 million seconds highlights potential outliers or misuse.

We have studied the usage patterns and ride duration trends. We now shift our focus to identifying the most popular and least popular start stations as well as end stations for bicycle usage in London.

### 4.3 What are the most popular and least popular start and end stations?

```python
# printing the most popular start and end station.
most_popular_start_station_query = """
SELECT start_station_name, COUNT(*) as count
FROM cycle_hires
GROUP BY start_station_name
ORDER BY count DESC
LIMIT 1
"""
most_popular_start_station = spark.sql(most_popular_start_station_query).first()

most_popular_end_station_query = """
SELECT end_station_name, COUNT(*) as count
FROM cycle_hires
GROUP BY end_station_name
ORDER BY count DESC
LIMIT 1
"""
most_popular_end_station = spark.sql(most_popular_end_station_query).first()
print(f"The most popular start station is: {most_popular_start_station['start_station_name']} with {most_popular_start_station['count']} rentals")
print(f"The most popular end station is: {most_popular_end_station['end_station_name']} with {most_popular_end_station['count']} rentals")
```

```python
# printing the least popular start and end station.
least_popular_start_station_query = """
SELECT start_station_name, COUNT(*) as count
FROM cycle_hires
GROUP BY start_station_name
ORDER BY count ASC
LIMIT 1
"""
least_popular_start_station = spark.sql(least_popular_start_station_query).first()

least_popular_end_station_query = """
SELECT end_station_name, COUNT(*) as count
FROM cycle_hires
GROUP BY end_station_name
ORDER BY count ASC
LIMIT 1
"""
least_popular_end_station = spark.sql(least_popular_end_station_query).first()

print(f"The least popular start station is: {least_popular_start_station['start_station_name']} with {least_popular_start_station['count']} rentals")
print(f"The least popular end station is: {least_popular_end_station['end_station_name']} with {least_popular_end_station['count']} rentals")

```

## 8. Machine Learning Approach
----------

The ML task chosen for this dataset was prediction of ride duration based on various features. The problem was converted into from a regression problem (prediction of duration value) to a classification problem i.e., prediction of category of ride based on duration. The ride categories were defined as 'Short', 'Medium', 'Long', and 'Very Long'. Features selected for the model include 'start_date', 'duration', 'bike_model', 'start_station_id', and 'end_station_id'.

Opting for a classification model over regression was driven by the nature of the ride duration data, which exhibited significant skewness and contained notable outliers as shown in Section 4.2. Classifiers inherently offer greater robustness against such outliers, ensuring that extreme values do not disproportionately impact model performance. Moreover, classification excels at unraveling the intricate, non-linear interactions that can exist between variables such as the time of day or the day of the week and the duration of the rideârelationships that might be oversimplified by a regression approach. From a business standpoint, the clarity of classification outcomes is particularly advantageous; focusing on categorical ride durationsâshort, medium, long, and very longâyields actionable insights that are directly aligned with operational and strategic objectives, as opposed to the more granular but less intuitive numerical predictions of exact durations.

### 8.1 : Data Preprocessing, Feature Engineering

The preprocessing and feature engineering steps for this classification problem are:

- Creating a new column ride_category based on ride duration.
- Extracting the day of the week and the hour of the day from the start_date to capture temporal patterns in ride usage.
- Converting categorical variables ('bike_model', 'start_station_id', 'end_station_id', and 'ride_category') into numerical values using StringIndexer and encoding them with OneHotEncoder for use in machine learning models.
- Assembling all the features, including one-hot encoded and continuous features, into a single vector column using VectorAssembler.


```python
ml_df=cycle_hire_df_clean.select('start_date','duration','bike_model', 'start_station_id', 'end_station_id')
ml_df = ml_df.withColumn(
    "ride_category",
    when(col("duration") < 300, "Short")
    .when((col("duration") >= 300) & (col("duration") < 1800), "Medium")
    .when((col("duration") >= 1800) & (col("duration") < 7200), "Long")
    .otherwise("Very Long")
)

#day of the week for the ride
ml_df = ml_df.withColumn('start_day_of_week', dayofweek('start_date'))

#start hour of the ride
ml_df = ml_df.withColumn('start_hour', hour('start_date'))
ml_df.show(5)
```

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.ml import Pipeline

#data pre-processing : preparing data for ML models
stringIndexer = StringIndexer(inputCols=['bike_model', 'start_station_id', 'end_station_id', 'ride_category'], outputCols=['bike_model_index', 'start_station_id_index', 'end_station_id_index', 'ride_category_index'],handleInvalid="keep")
oneHotEncoder = OneHotEncoder(inputCols=['bike_model_index', 'start_station_id_index', 'end_station_id_index', 'ride_category_index'], outputCols=['bike_model_vec', 'start_station_id_vec', 'end_station_id_vec', 'ride_category_vec'])

assembler = VectorAssembler(inputCols=['start_day_of_week', 'start_hour', 'bike_model_vec', 'start_station_id_vec', 'end_station_id_vec'], outputCol='features')
```

### 8.2 : Model Training

```python
#train-test split
(trainingData, testData) = ml_df.randomSplit([0.8, 0.2], seed=843)
```

```python
# importing multiclassClassifierEvalutor
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
evaluator_classification = MulticlassClassificationEvaluator(labelCol="ride_category_index", predictionCol="prediction")
```

```python
# logistic regression for classification
from pyspark.ml.classification import LogisticRegression
logisticReg = LogisticRegression(featuresCol='features', labelCol='ride_category_index')
pipeline_log = Pipeline(stages=[stringIndexer, oneHotEncoder, assembler, logisticReg])
model_log = pipeline_log.fit(trainingData)
predictions_log = model_log.transform(testData)
accuracy_log = evaluator_classification.evaluate(predictions_log)
f1_log = evaluator_classification.setMetricName("f1").evaluate(predictions_log)
precision_log = evaluator_classification.setMetricName("weightedPrecision").evaluate(predictions_log)
recall_log = evaluator_classification.setMetricName("weightedRecall").evaluate(predictions_log)

print(f"Accuracy: {accuracy_log}")
print(f"Precision: {precision_log}")
print(f"Recall: {recall_log}")
print(f"F1 Score: {f1_log}")
```

```python
# logistic regression with regularization
logisticReg1 = LogisticRegression(featuresCol='features', labelCol='ride_category_index',regParam = 0.1)
pipeline_log1 = Pipeline(stages=[stringIndexer, oneHotEncoder, assembler, logisticReg1])
model_log1 = pipeline_log1.fit(trainingData)
predictions_log1 = model_log1.transform(testData)
accuracy_log1 = evaluator_classification.evaluate(predictions_log1)
f1_log1 = evaluator_classification.setMetricName("f1").evaluate(predictions_log1)
precision_log1 = evaluator_classification.setMetricName("weightedPrecision").evaluate(predictions_log1)
recall_log1 = evaluator_classification.setMetricName("weightedRecall").evaluate(predictions_log1)

print(f"Accuracy: {accuracy_log1}")
print(f"Precision: {precision_log1}")
print(f"Recall: {recall_log1}")
print(f"F1 Score: {f1_log1}")
```

```python
# logistic regression with regularization
logisticReg2 = LogisticRegression(featuresCol='features', labelCol='ride_category_index',regParam = 0.01)
pipeline_log2 = Pipeline(stages=[stringIndexer, oneHotEncoder, assembler, logisticReg2])
model_log2 = pipeline_log2.fit(trainingData)
predictions_log2 = model_log2.transform(testData)
accuracy_log2 = evaluator_classification.evaluate(predictions_log2)
f1_log2 = evaluator_classification.setMetricName("f1").evaluate(predictions_log2)
precision_log2 = evaluator_classification.setMetricName("weightedPrecision").evaluate(predictions_log2)
recall_log2 = evaluator_classification.setMetricName("weightedRecall").evaluate(predictions_log2)

print(f"Accuracy: {accuracy_log2}")
print(f"Precision: {precision_log2}")
print(f"Recall: {recall_log2}")
print(f"F1 Score: {f1_log2}")
```

### 8.3 : Plot 15 - Model Performance Evaluation

```python
# Comparative analysis of the model performance
metrics = {
    "Accuracy": [0.714822579240802, 0.7106900083535161, 0.7135245450763438],
    "Precision": [0.7427963565089593, 0.7674609419869375, 0.7463232078269669],
    "Recall": [0.8002576101206942, 0.7996670424184845, 0.7463232078269669],
    "F1 Score": [0.714822579240802, 0.7106900083535161, 0.7463232078269669]
}

labels = list(metrics.keys())
model1 = [metrics[label][0] for label in labels]
model2 = [metrics[label][1] for label in labels]
model3 = [metrics[label][2] for label in labels]

x = np.arange(len(labels))
width = 0.25

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width, model1, width, label='Regularization (regParam=0.1)')
rects2 = ax.bar(x, model2, width, label='No Regularization')
rects3 = ax.bar(x + width, model3, width, label='Regularization (regParam=0.01)')

ax.set_ylabel('Value')
ax.set_xlabel('Evaluation Metric')
ax.set_title('Comparison of Logistic Regression Models')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

min_value = min(min(model1), min(model2), min(model3))
max_value = max(max(model1), max(model2), max(model3))
buffer = (max_value - min_value) * 0.05
ax.set_ylim(min_value - buffer, max_value + buffer)

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.4f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
fig.tight_layout()
plt.show()
```

From the plot, we observe that the model with regularization (regParam=0.1) generally performs slightly better in terms of Recall, but closely matches the F1 Score of the model with a lower regularization parameter (regParam=0.01). The model without regularization shows a slight improvement in Precision, potentially indicating overfitting to the training data. The Recall values are notably consistent across models, suggesting that the regularization parameter primarily impacts the balance between Precision and F1 Score, optimizing for overall error reduction without significantly sacrificing recall. This indicates that adjusting the regularization strength can help in fine-tuning the model's performance to achieve a desired balance between error types.
